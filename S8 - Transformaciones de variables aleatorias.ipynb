# 8 - Transformaciones de variables aleatorias

> Las transformaciones permiten determinar la función de densidad probabilística de una variable aleatoria $Y$ a partir del conocimiento de otra variable aleatoria $X$, relacionadas por la **función** $Y = T(X) = g(X)$. Es común en situaciones donde el resultado de un proceso o la salida de un sistema depende de una entrada aleatoria.

## 8.1 - Simulación de Montecarlo para la corriente y la potencia

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import csv

# Resistencia
R = 10

# Repeticiones
N = 5000

# Distribución de la corriente
I = stats.uniform(1, 2)

# Inicialización de los valores de corriente y potencia (vector de ceros)
Irvs = [0] * N
P = [0] * N

# "Simulación"
for i in range(N):
    Irvs[i] = I.rvs()
    P[i] = Irvs[i]**2 * R
    
# Crear histograma de I
a, b, c = plt.hist(Irvs, 20, density=True)
plt.xlabel('Corriente')
plt.ylabel('Densidad')
plt.show()

# Crear histograma de P
a, b, c = plt.hist(P, 20, density=True)
plt.xlabel('Potencia')
plt.ylabel('Densidad')
plt.show()
```

## 8.2 - Transformación cuadrática de una variable aleatoria normal

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import csv

# Repeticiones
N = 500

# Distribución de X
Xva = stats.truncnorm(-2,2)

# Función aplicada sobre X
X = Xva.rvs(N)
Y = X**2

# Crear histograma de X
a, b, c = plt.hist(X, 20, density=True)
plt.xlabel('Variable')
plt.ylabel('Densidad')
plt.show()

# Crear histograma de Y
a, b, c = plt.hist(Y, 20, density=True)
plt.xlabel('Variable')
plt.ylabel('Densidad')
plt.show()

# Crear archivo
with open('cuadratica.csv', 'w', newline='') as archivo:
    escritura = csv.writer(archivo)
    escritura.writerow(['x','y'])
    for i in range(N):
        escritura.writerow([X[i],Y[i]])
```
